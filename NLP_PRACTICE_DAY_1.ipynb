upper case to lower case


```python
lower_text_1=[]
def t0_lower(data):
  for words in raw_text:
    lower_text_1.append(str.lower(words))
```

```python
t0_lower(raw_text)
lower_text_1
```

TOKENIZATION

```python
toekn_text_1=[]
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
sent_tok=[]
for sent in lower_text_1:
  sent=sent_tokenize(sent)
  sent_tok.append(sent)
```

```python
sent_tok
```

word and sentence tokenization

```python
#word tokenize
token_text_1=[word_tokenize(i) for i in lower_text_1]
token_text_1
```

```python
#token text is
token_text_1
```

REMOVING_PUNCHUATIONS


```python
import re
nospcl_text_1=[]

for words in token_text_1:
  clean=[]
  for w in words:
    res=re.sub(r'[^\w\s]',"",w)
    if res!="":
      clean.append(res)
    nospcl_text_1.append(clean)
```

```python
nospcl_text_1
```

REMOVE STOP WORDS


```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_word_text_1=[]
for words in nospcl_text_1:
  w=[]
  for word in words:
    if word not in stopwords.words('english'):
      w.append(word)
    stop_word_text_1.append(w)
```

```python
stop_word_text_1
```

5.STEMMING  // CONVERTING THE WORD INTO ROOT FORM

```python
from nltk.stem.porter import  PorterStemmer
```

```python
port=PorterStemmer()
```

```python
a=[port.stem(i) for i in ["reading","washing","wash","driving"]]
a
```

```python
stem_text_1=[]
for words in stop_word_text_1:
  w=[]
  for word in words:
    w.append(port.stem(word))
  stem_text_1.append(w)
```

```python
stem_text_1
```

LEMMATIZATION

```python
from nltk.stem import WordNetLemmatizer
wnet=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
lem_text=[]
for words in stop_word_text_1:
  w=[]
  for word in words:
    w.append(wnet.lemmatize(word))
  lem_text.append(w)
```
